```python
import matplotlib.pyplot as plt
%matplotlib inline

import os
import numpy as np

import wfdb
```

# read record

```python
record = wfdb.rdrecord('100', sampto = 3000, pb_dir="mitdb")
annotation = wfdb.rdann('100', 'atr', sampto = 3000, pb_dir="mitdb")

wfdb.plot_wfdb(record=record, annotation=annotation,
               title='Record 100 from MIT-BIH Arrhythmia Database',
               time_units='seconds')
```
